```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('imdb-top-1000.csv')
df.head()
```

```python
genre_group=df.groupby('Genre')
```

```python
genre_earnings = genre_group['Gross'].sum().sort_values(ascending=False)
```

```python
print(genre_earnings.head(3))
```

```python
avg_rating = genre_group['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
print(avg_rating.head(1))
```

```python
director=df.groupby('Director')
popularity=director['No_of_Votes'].sum().sort_values(ascending=False)
print(popularity.head(1))
```

```python
df_sort = df.sort_values('IMDB_Rating', ascending=False)
df_sort = df.groupby('Genre').first()
```

```python
df_sort = df_sort.reset_index() 
result = df_sort[['Genre', 'Series_Title', 'IMDB_Rating']]
print(result)
```
