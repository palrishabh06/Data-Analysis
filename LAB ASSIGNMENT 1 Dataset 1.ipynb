```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('student_health_lifestyle_dataset.csv')
df.head()
```

# Locate missing values

```python
df.isnull().sum()
```

# Fill the missing values using Mean, Median , Mode and Interpolation

```python
df['Calories_Intake'] = df['Calories_Intake'].interpolate()
df['Sleep_Hours'].fillna(df['Sleep_Hours'].mean(), inplace=True)
df['BMI'].fillna(df['BMI'].mean(), inplace=True)
df['Exercise_Hours'].fillna(df['Exercise_Hours'].median(), inplace=True)
```

```python
df.head()
```

# Check Duplicates

```python
df.duplicated().sum()
```

# Unique values

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].replace({'female':'Female','Femlae': 'Female', 'femlae': 'Female'})
```

```python
df['Gender'].unique()
```

# Textual data uniform (Gender -> Lower Case)

```python
df['Gender'] = df['Gender'].str.lower()
df.head()
```

# Identify and remove the outliers

```python
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['BMI'] >= Q1 - 1.5 * IQR) & (df['BMI'] <= Q3 + 1.5 * IQR)]
df.tail(20)
```

# Perform data Normalization
    using MinMax scaler on Sleep Hours, Calories Intake, BMI, & Exercise Hours.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = MinMaxScaler()
df[['Sleep_Hours','Calories_Intake','BMI','Exercise_Hours']] = scaler.fit_transform(
    df[['Sleep_Hours','Calories_Intake','BMI','Exercise_Hours']]
)
df.head()
```

# Perform data Standardization
    using Standard scaler on Screen time & Exam score.

```python
scaler = StandardScaler()
df[['Exam_Score','Screen_Time']] = scaler.fit_transform(
    df[['Exam_Score','Screen_Time']]
)
```
